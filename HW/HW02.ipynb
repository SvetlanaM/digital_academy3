```python
"""
Zmente defaultný pandas index na index stĺpca id
"""
coursesDf = coursesDf.set_index("id")
```

```python
"""
Vypíšte si dáta z labelmi od 50 po 55 nad stĺpcami couches, course_price, notes
"""
coursesDf.loc[50:55, ["couches", "course_price", "notes"]]
```

```python
"""
Upravte všetky mená koučov s stĺpci couches na veľké písmená a uložte do daného stĺpca
"""
coursesDf["couches"] = coursesDf["couches"].str.upper()
coursesDf.head()
```

```python
"""
Vypíšte počet prázdnych hodnôt zo stĺpca max_capacity
"""
coursesDf["max_capacity"].isnull().sum()
```

```python
"""
Nahraďte v stĺpci max_capacity prázdne hodnoty, hodnotou 0 a nastavte ako int dátový typ
"""
coursesDf["max_capacity"] = coursesDf["max_capacity"].fillna(1).astype(np.int)
coursesDf.head()
```

```python
"""
Vypíšte kurzy, kde lektorom je Pavel Lasák
"""
coursesDf[coursesDf["couches"] == 'PAVEL LASÁK']
```

```python
"""
Vypíšte kurzy, ktoré začínajú  v rozmedzí dátumov od 01. 12. 2016 do 31. 12. 2016
"""
coursesDf[["title", "course_start_date"]][(pd.to_datetime(coursesDf["course_start_date"]).dt.month.between(12,12))]
```

```python
coursesDf[["title", "course_start_date"]][(pd.to_datetime(coursesDf["course_start_date"]).dt.month == 12)]
```

```python
"""
Vypíšte kurzy, ktoré sú drahšie ako 1500 Kč a nemajú žiadneho lektora
"""
coursesDf[(coursesDf["course_price"] > 1500) & (coursesDf["couches"] == '')]
```

```python
"""
Vypíšte kurzy, kde cena je 300 Kč alebo kurzy, kde koučom je Pavel Ungr
"""
coursesDf[(coursesDf["course_price"] == 300) | (coursesDf["couches"] == 'PAVEL UNGR')]
```

```python
"""
V stĺpci publish a open registration nahradte True za 1 a False za 0 a uložte ako dátový typ int
"""
coursesDf[["publish", "open_registration"]] = coursesDf[["publish", "open_registration"]].astype(np.int)
coursesDf.head()
```

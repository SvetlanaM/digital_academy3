```python
import pandas as pd
```

```python
coursesDf = pd.read_json('https://czechitas-app.herokuapp.com/api/v1/courses/open/')
```

```python
"""
2. Očiste si dátový set o kurzy, ktoré nemajú kouča a uložte do novej premennej s názvom courses
"""
coursesDf = coursesDf[coursesDf["couches"] != '']
```

```python
"""
3. Spravte pivotnú/kontingenčnú tabuľku, kde budú dáta za koučov a celková suma za ich kurz . 
Zároveň zrátajte sumu za všetky kurzy pod All. 
"""
pivot = pd.pivot_table(coursesDf, index='couches', margins = True, values='course_price', aggfunc='sum')
pivot
```

```python
"""
4. Zrátajte počet kurzov lektorov cez groupby
"""
coursesDf.groupby('couches')["couches"].count()
```

```python
"""
5. Zgrupte hodnoty za koučov, dátum štartu kurzu a zrátajte sumu za daný kurz 
"""
coursesDf.groupby(['couches', 'course_start_date'])["course_price"].agg(sum)
```

```python
"""
6. Nad zgrupením z úlohy vyššie použite describe() 
"""
coursesDf.groupby(['couches', 'course_start_date'])["course_price"].describe()
```

```python
"""
7. Nad zgrupenim z ulohy 5 pouzite nakonci este metodu unstack() 
"""
coursesDf.groupby(['couches', 'course_start_date'])["course_price"].agg(sum).unstack()
```

```python
"""
8. Z datoveho setu odstrante stlpce contact_person a course_category
"""
coursesDf.drop(['contact_person', 'course_category'], axis = 1, inplace = True)
coursesDf.head()
```

```python
"""
10. Napíš program, ktorý vo vete "Ahoj, ja som Harry Potter a toto je mojich 12 potkanov." 
zráta počet znakov a počet čísel.
"""
sentence = input("Enter the sentence: ")
my_dict = {"DIGITS" : 0, "LETTERS" : 0}
for s in sentence:
    if s.isdigit():
        my_dict["DIGITS"]+=1
    elif s.isalpha():
        my_dict["LETTERS"]+=1
    else:
        continue
print ("LETTERS IS: ", my_dict["LETTERS"])
print ("DIGITS IS: ", my_dict["DIGITS"])
```

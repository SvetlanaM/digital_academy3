```python
df = pd.read_csv('csv/fit.recepty.csv', sep=',')
df.reset_index(inplace=True)
df.head(11)
```

```python
import re
results = pd.Series(df["caption"]).str.extractall('(\d+[\w*[" "]*\w*)')
results.reset_index(inplace=True)
results
```

```python
results1 = pd.Series(df["caption"]).str.extractall('^([\w*[" "]*[@])')
results1.reset_index(inplace=True)
results1
```

```python
merged_inner = pd.merge(left=results1,right=df, left_on='level_0', right_on='index')
merged_inner = merged_inner.drop(['level_0', 'match', 'code'], axis = 1)
merged_inner.head(11)
```

```python
merged_left = pd.merge(left=results,right=merged_inner, left_on='level_0', right_on='index')
merged_left.head(11)
```

```python
import datetime 
def convert_date(date):
    new_date = datetime.datetime.fromtimestamp(int(date))
    return new_date
date = merged_left["date"]
date = date.apply(convert_date)
merged_left["date"] = date
merged_left.head(11)
```

```python
merged_left.to_csv('csv/final_dataset.csv')
```
